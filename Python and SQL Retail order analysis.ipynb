#### To change the unknown and not available data to nan

```python
df = pd.read_csv("retail_orders_data/orders.csv",na_values=['Not Available', 'unknown']) 
df['Ship Mode'].unique()
```

```python
df.columns
```

#### Change the column name to lower for easier usecase and also use '_' in the place of spaces 

```python
df.columns=df.columns.str.lower()
```

```python
df.columns
```

```python
df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df.head(5)
```

#### Creating new column discount,sale_price,profit

```python
df['discount']=df['list_price']*df['discount_percent']*.01
df['discount']
```

```python
df['sale_price']=df['list_price']-df['discount']
df['sale_price']
```

```python
df['profit']=df['sale_price']-df['cost_price']
df['profit']
```

#### order date is in the object format changing into datetime

```python
df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%m-%d")
```

```python
df.dtypes
```

#### Droppping the unwanted columns 

```python
df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
```

```python
import sqlalchemy as sal
import urllib

# Define connection parameters
server = "YOUR_SERVER_NAME"         # e.g., 'SERVER\\SQLEXPRESS'
database = "YOUR_DATABASE_NAME"     # e.g., 'master'
driver = "ODBC Driver 17 for SQL Server"

# Use Windows Authentication
connection_string = f"mssql+pyodbc://{server}/{database}?driver={urllib.parse.quote_plus(driver)}&trusted_connection=yes"

# Create engine and connect
engine = sal.create_engine(connection_string)
conn = engine.connect()
print("✅ Connection successful!")
```

```python
df.to_sql('df_orders', con=conn, index=False, if_exists='append')
```
